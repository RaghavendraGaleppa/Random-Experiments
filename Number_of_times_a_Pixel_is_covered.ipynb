### In this notebook, I will see how many times does a particular kernel of any size, covers all the pixels in the given image size

```
import numpy as np
import matplotlib.pyplot as plt

def image_per_pixel_strength(img_size=(224,224,3), kernel_size=3):
    img = np.zeros(shape=img_size, dtype=np.float)
    for i in range(img_size[1] - kernel_size + 1):
        for j in range(img_size[0] - kernel_size + 1):
            img[i:i+3, j:j+3] += 0.1

    plt.figure(figsize=(10,10))
    plt.imshow(img)
```

```
# The center part of the image is reference the most amount of times
image_per_pixel_strength((5,5,3))
```

```
image_per_pixel_strength((10,10,3))
```

```

image_per_pixel_strength((20,20,3))
```
