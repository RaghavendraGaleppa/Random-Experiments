```
import matplotlib.pyplot as plt
import torch.nn.functional as F
import cv2

def plot_2_tensors(t1, t2):
    t1 = t1.numpy().squeeze().transpose(1,2,0)
    t2 = t2.numpy().squeeze().transpose(1,2,0)

    fig, ax = plt.subplots(1,2,figsize=(15,15))
    ax[0].imshow(t1)
    ax[1].imshow(t2)

def mp_v_ap(img,k_size=2):
    t1 = torch.from_numpy(img.transpose(2,0,1)).unsqueeze(0)
    t1 = F.max_pool2d(t1, k_size)

    t2 = torch.from_numpy(img.transpose(2,0,1)).unsqueeze(0)
    t2 = F.avg_pool2d(t2, k_size)

    plot_2_tensors(t1,t2)
```

```
img = cv2.imread('pic.jpg')[...,::-1]/255.0
mp_v_ap(img,2)
```
